```python
from collections import deque

def print_puzzle(puzzle):
    for row in puzzle:
        print(' '.join(row))
    print()

def get_possible_moves(puzzle):

    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == '_':
                x, y = i, j

    moves = [(x-1, y), (x+1, y), (x, y-1), (x, y+1)] 
    valid_moves = [(i, j) for i, j in moves if 0 <= i < 3 and 0 <= j < 3]
    return valid_moves, x, y

def bfs(puzzle, goal):
    
    queue = deque([puzzle])
    visited = set([str(puzzle)])

    while queue:
        puzzle = queue.popleft()
        print_puzzle(puzzle)

        if puzzle == goal:
            return True

        moves, x, y = get_possible_moves(puzzle)
        for move in moves:
            new_puzzle = [row.copy() for row in puzzle]
            new_puzzle[x][y], new_puzzle[move[0]][move[1]] = new_puzzle[move[0]][move[1]], new_puzzle[x][y]
            str_puzzle = str(new_puzzle)
            if str_puzzle not in visited:
                visited.add(str_puzzle)
                queue.append(new_puzzle)

    return False

initial_puzzle = [['_', '1', '3'], ['4', '2', '5'], ['7', '8', '6']]
goal_puzzle = [['1', '2', '3'], ['4', '5', '6'], ['7', '8', '_']]

if bfs(initial_puzzle, goal_puzzle):
    print("Solution found.")
else:
    print("No solution found.")
```
