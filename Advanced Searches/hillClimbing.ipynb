```python
import random

# Define the objective function
def objective_function(x):
    return -x**2 + 4*x + 10

# Define the hill climbing algorithm
def hill_climbing(objective_function, initial_solution, step_size, max_iterations):
    current_solution = initial_solution
    current_value = objective_function(current_solution)

    for iteration in range(max_iterations):
        # Generate a neighbor solution
        neighbor_solution = current_solution + random.uniform(-step_size, step_size)
        neighbor_value = objective_function(neighbor_solution)

        # If the neighbor is better, move to the neighbor
        if neighbor_value > current_value:
            current_solution = neighbor_solution
            current_value = neighbor_value

        # Print the current state
        print(f"Iteration {iteration}: Current solution = {current_solution}, Value = {current_value}")

    return current_solution, current_value

# Initial solution, step size, and number of iterations
initial_solution = random.uniform(-10, 10)
step_size = 0.1
max_iterations = 100

# Run the hill climbing algorithm
best_solution, best_value = hill_climbing(objective_function, initial_solution, step_size, max_iterations)

print(f"Best solution: {best_solution}, Best value: {best_value}")
```

```python
import random



def hcs(states, budget, iters):

    curr = random.choice(list(states.keys()))
    total = states[curr]['cost']
    i = 0
    solutions = [curr]

    while i < iters and total <= budget:

        neighbor = random.choice(list(states.keys()))

        if states[neighbor]['rating'] > states[curr]['rating'] and ( states[neighbor]['cost'] + states[curr]['cost'] ) <= budget:
            curr = neighbor
            total += states[neighbor]['cost']
            solutions.append(neighbor)

        i += 1

    return solutions







budget = 10000
states = {
    "feature1": {"cost": 100, "rating": 3.5},
    "feature2": {"cost": 200, "rating": 4.2},
    "feature3": {"cost": 150, "rating": 4.0},
    "feature4": {"cost": 300, "rating": 3.8},
    "feature5": {"cost": 250, "rating": 4.5},
    "feature6": {"cost": 350, "rating": 3.6}
}

s = hcs(states, budget, 1000)
s
```

```python
import random

class EightQueens:
    def __init__(self, size=8):
        self.size = size
        self.board = [random.randint(0, size - 1) for _ in range(size)]
    
    def fitness(self, board):
        non_attacking = 0
        for i in range(self.size):
            for j in range(i + 1, self.size):
                if board[i] != board[j] and abs(board[i] - board[j]) != j - i:
                    non_attacking += 1
        return non_attacking

    def get_neighbors(self, board):
        neighbors = []
        for i in range(self.size):
            for j in range(self.size):
                if board[i] != j:
                    neighbor = list(board)
                    neighbor[i] = j
                    neighbors.append(neighbor)
        return neighbors
    
    def hill_climbing(self):
        current_board = self.board
        current_fitness = self.fitness(current_board)
        while True:
            neighbors = self.get_neighbors(current_board)
            best_neighbor = None
            best_neighbor_fitness = -1
            for neighbor in neighbors:
                fitness = self.fitness(neighbor)
                if fitness > best_neighbor_fitness:
                    best_neighbor_fitness = fitness
                    best_neighbor = neighbor
            
            if best_neighbor_fitness <= current_fitness:
                break
            current_board = best_neighbor
            current_fitness = best_neighbor_fitness
            print(f"Current fitness: {current_fitness}")
        
        return current_board, current_fitness
    
    def print_board(self, board):
        for row in range(self.size):
            line = ""
            for col in range(self.size):
                if board[col] == row:
                    line += "Q "
                else:
                    line += ". "
            print(line)
        print("\n")

# Initialize the EightQueens problem
eight_queens = EightQueens()

# Solve the problem using hill climbing
solution, fitness = eight_queens.hill_climbing()

# Print the solution
print(f"Solution fitness: {fitness}")
eight_queens.print_board(solution)
```
