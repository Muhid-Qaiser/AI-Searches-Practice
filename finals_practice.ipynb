## Genetic Algo

```python
# knapsack problem
import random

class knapsack:

    def __init__(self, values, weights, pop_size, chromosome_size, mutation_rate, max_weight, gen):

        self.m_rate = mutation_rate
        self.max_weight = max_weight
        self.values = values
        self.weights = weights
        self.gen = gen
        self.pop_size = pop_size
        self.chromosome_size = chromosome_size

        temp = [ [random.choice([0, 1]) for _ in range(chromosome_size)] for _ in range(pop_size) ]
        self.population = [ [chromosome, self.fitness_func(chromosome)] for chromosome in temp ]
        self.population.sort(key=lambda x: x[1], reverse=True)

    
    def fitness_func(self, chromosome):

        fitness = 0
        weight = 0
        for index in range(len(chromosome)):
            if chromosome[index] == 1:
                fitness += self.values[index]
                weight += self.weights[index]

        if weight > self.max_weight:
            return 0
        return fitness


    def parent_selection(self, population):
        
        fitness_list = [ fitness for chromosome, fitness in population  ]
        total_fitness = sum(fitness_list)
        fitness_percentage = [ fitness / total_fitness for fitness in fitness_list]
        
        parent1, parent2 = random.choices(population, fitness_percentage, k=2)

        return parent1, parent2


    def cross_over(self, parent1, parent2):

      crossover_point = random.randint(0, self.chromosome_size - 1)
      child1 = parent1[:crossover_point] + parent2[crossover_point:]
      child2 = parent2[:crossover_point] + parent1[crossover_point:]

      return child1, child2


    def mutation(self, chromosome):

      mutated_sol = []

      for i in chromosome:
        if random.random() < self.m_rate:
          mutated_sol.append(1 - chromosome[i])
        else:
          mutated_sol.append(chromosome[i])

      return mutated_sol

        

    def genetic_algorithm(self):

        for i in range(self.gen):

            for _ in range(self.pop_size):

                parent1, parent2 = self.parent_selection(self.population)
                child1, child2 = self.cross_over(parent1[0], parent2[0]) 
                child1 = self.mutation(child1)
                child2 = self.mutation(child2)


                self.population.append( (child1, self.fitness_func(child1)) )
                self.population.append( (child2, self.fitness_func(child2)) )

                self.population.sort(key=lambda x: x[1], reverse=True)
                self.population = self.population[:self.pop_size]

            print(f"Best in Gen {i}: ", self.population[0])






values = [14, 23, 8, 9, 17, 15]
weights = [1, 3, 7, 4, 5, 1]
k = knapsack(values, weights, 6, 6, 0.1,10,10)
k.genetic_algorithm()

```
